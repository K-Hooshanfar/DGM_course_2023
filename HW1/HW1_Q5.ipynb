```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("ngaussian (1).csv" , delimiter=' ',header=None)
```

```python
df
```

```python
X1=df[0].values.reshape(-1, 1)  
X2=df[1].values.reshape(-1, 1)  
```

```python
plt.scatter(X1, X2 , label='Original Data')
plt.legend()
plt.xlabel('X1')
plt.ylabel('X2')
plt.title("X1 vs X2")
plt.show()
```

SCM M1

```python
model = LinearRegression(fit_intercept=False)

# Fit the model to the data
model.fit(X1, X2)

intercept = model.intercept_  
slope12 = model.coef_[0]  

print(f"B12 is: {slope12}")
```

```python
plt.scatter(X1, X2-slope12*X1)
plt.xlabel('X1')
plt.ylabel('X2-slope12*X1')
plt.show()
```

SCM M2

```python
model = LinearRegression(fit_intercept=False)

# Fit the model to the data
model.fit(X2, X1)

intercept = model.intercept_  
slope21 = model.coef_[0]  

print(f"B21 is: {slope21}")
```

```python
plt.scatter(X2, X1-slope21*X2)
plt.xlabel('X2')
plt.ylabel('X1-slope21*X2')
plt.show()
```
